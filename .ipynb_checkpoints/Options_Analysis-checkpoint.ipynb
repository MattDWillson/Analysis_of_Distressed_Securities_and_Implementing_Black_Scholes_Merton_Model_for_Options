```python
#Implementing Black Scholes Time Series Analysis of Hilton

#Define Variables

#Interest Rate
r = .01

#Underlying Stock Price
S_HLT = HLT_Closing_Price_Data["HLT"]

#Strike Price
K_HLT = 1.05*S_HLT

#Time to Expiration
T_HLT = 60/365

#Volatility or Standard Deviation
sigma_HLT = .20

def blackscholes(r, S, K, T, sigma, type="C"):
    "Calculate BS option price for a call/put"
    d1 = (np.log(S/K) + (r + sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    try:
        if type == "C":
            price = S*norm.cdf(d1, 0, 1) - K*np.exp(-r*T)*norm.cdf(d2, 0, 1)
        elif type == "P":
            price = K*np.exp(-r*T)*norm.cdf(-d2, 0, 1) - S*norm.cdf(-d1, 0, 1)
        return price
    except:
        print("Please confirm all option parameters above!")
```

```python
HLT_call_option_df = pd.DataFrame(round(blackscholes(r, S_HLT, K_HLT, T_HLT, sigma_HLT, type="C"), 2) )
HLT_call_option_df.head()
```

```python
# To calculate IV
import mibian
#Use the mibian library to calculate the implied volatility.
#The syntax for the variable values is in the format as mentioned below :
#BS([UnderlyingPrice, StrikePrice, InterestRate, Daystoexpiration],callPrice=x)

#Calculate IV using the mibian library for 30 day expiry NIFTY 50 call options (Strike at 1.05X underlying value)

Oct_2019_NIFTY = mibian.BS([11359, 11950, 1, 30], callPrice=35)
Nov_2019_NIFTY = mibian.BS([11787, 12400, 1, 30], callPrice=29)
Dec_2019_NIFTY = mibian.BS([12039, 12650, 1, 30], callPrice=34)
Jan_2020_NIFTY = mibian.BS([12255, 12850, 1, 30], callPrice=8)
Feb_2020_NIFTY = mibian.BS([12056, 12650, 1, 30], callPrice=44)
Mar_2020_NIFTY = mibian.BS([11202, 11750, 1, 30], callPrice=56)
Apr_2020_NIFTY = mibian.BS([8281, 8700, 1, 30], callPrice=469)
May_2020_NIFTY = mibian.BS([9380, 9750, 1, 30], callPrice=446)
Jun_2020_NIFTY = mibian.BS([9029, 9500, 1, 30], callPrice=101)
Jul_2020_NIFTY = mibian.BS([10552, 11100, 1, 30], callPrice=65)
Aug_2020_NIFTY = mibian.BS([11073, 11600, 1, 30], callPrice=77)
Sep_2020_NIFTY = mibian.BS([11647, 12200, 1, 30], callPrice=28.35)
Oct_2020_NIFTY = mibian.BS([11559, 12050, 1, 30], callPrice=27)


#Create Implied Volatility Dict

Oct_2019_IV = Oct_2019_NIFTY.impliedVolatility
Nov_2019_IV = Nov_2019_NIFTY.impliedVolatility
Dec_2019_IV = Dec_2019_NIFTY.impliedVolatility
Jan_2020_IV = Jan_2020_NIFTY.impliedVolatility
Feb_2020_IV = Feb_2020_NIFTY.impliedVolatility
Mar_2020_IV = Mar_2020_NIFTY.impliedVolatility
Apr_2020_IV = Apr_2020_NIFTY.impliedVolatility
May_2020_IV = May_2020_NIFTY.impliedVolatility
Jun_2020_IV = Jun_2020_NIFTY.impliedVolatility
Jul_2020_IV = Jul_2020_NIFTY.impliedVolatility
Aug_2020_IV = Aug_2020_NIFTY.impliedVolatility
Sep_2020_IV = Sep_2020_NIFTY.impliedVolatility
Oct_2020_IV = Oct_2020_NIFTY.impliedVolatility

IV_Dict = {
    "October Call": Oct_2019_IV,
    "November Call": Nov_2019_IV,
    "December Call": Dec_2019_IV,
    "January Call": Jan_2020_IV, 
    "February Call": Feb_2020_IV, 
    "March Call": Mar_2020_IV, 
    "April Call": Apr_2020_IV, 
    "May Call": May_2020_IV, 
    "June Call": Jun_2020_IV,
    "July Call": Jul_2020_IV,
    "August Call": Aug_2020_IV,
    "September Call": Sep_2020_IV,
    "October Call": Oct_2020_IV}

IV_df = pd.DataFrame.from_dict(IV_Dict, orient='index', columns=['NIFTY 30 Day Call Option'])
IV_df
```
