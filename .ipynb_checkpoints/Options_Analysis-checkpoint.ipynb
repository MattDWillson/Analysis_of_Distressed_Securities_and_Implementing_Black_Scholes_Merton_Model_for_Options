```python
#Implementing Black Scholes Time Series Analysis of Hilton

#Define Variables

#Interest Rate
r = .01

#Underlying Stock Price
S_HLT = HLT_Closing_Price_Data["HLT"]

#Strike Price
K_HLT = 1.05*S_HLT

#Time to Expiration
T_HLT = 60/365

#Volatility or Standard Deviation
sigma_HLT = .20

def blackscholes(r, S, K, T, sigma, type="C"):
    "Calculate BS option price for a call/put"
    d1 = (np.log(S/K) + (r + sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    try:
        if type == "C":
            price = S*norm.cdf(d1, 0, 1) - K*np.exp(-r*T)*norm.cdf(d2, 0, 1)
        elif type == "P":
            price = K*np.exp(-r*T)*norm.cdf(-d2, 0, 1) - S*norm.cdf(-d1, 0, 1)
        return price
    except:
        print("Please confirm all option parameters above!")
print(blackscholes(r, S_HLT, K_HLT, T_HLT, sigma_HLT, type="C"))
```

```python
HLT_call_option_df = pd.DataFrame(round(blackscholes(r, S_HLT, K_HLT, T_HLT, sigma_HLT, type="C"), 2) )
HLT_call_option_df
```

```python
# To calculate IV
import mibian
#Use the mibian library to calculate the implied volatility.
##The syntax for the variable values is in the format as mentioned below :
#BS([UnderlyingPrice, StrikePrice, InterestRate, Daystoexpiration],callPrice=x)


#Calculate IV for near the money NIFTY 50 in Dec 2019

Dec_26_NIFTY = mibian.BS([12048, 12100, 1, 30], callPrice=147)
Dec_26_NIFTY.impliedVolatility
```

```python
#Create Implied Volatility Dict

Dec_2019_IV = Dec_26_NIFTY.impliedVolatility
Jan_2020_IV = Jan_30_NIFTY.impliedVolatility
Feb_2020_IV = Feb_27_NIFTY.impliedVolatility
Mar_2020_IV = Mar_27_NIFTY.impliedVolatility
Apr_2020_IV = Apr_27_NIFTY.impliedVolatility
May_2020_IV = May_27_NIFTY.impliedVolatility
Jun_2020_IV = Jun_27_NIFTY.impliedVolatility

IV_Dict = {
    "December Call": Dec_2019_IV,
    "January Call": Jan_2020_IV, 
    "February Call": Feb_2020_IV, 
    "March Call": Mar_2020_IV, 
    "April Call": Apr_2020_IV, 
    "May Call": May_2020_IV, 
    "June Call": Jun_2020_IV}

IV_df = pd.DataFrame.from_dict(IV_Dict, orient='index', columns=['NIFTY 30 Day Call Option'])
IV_df
```

```python
#Implied Volatility Time Series 

ax = IV_df.plot(figsize=(10,10))

ax.set_ylabel("Implied Volatility", fontsize = 12)
ax.set_xlabel("Option Expiry - End of Month", fontsize = 12)
ax.set_title("Fluctuation of Implied Volatility for NIFTY 50 Call Options", fontsize = 14)
```
