```python
#Implementing Black Scholes formula for Hilton

#Define Variables

#Interest Rate
r = .01

#Udnerlying Stock Price
S = 125.85

#Strike Price
K = 130

#Time to Expiration
T = 30/365

#Volatility or Standard Deviation
sigma = .20

def blackscholes(r, S, K, T, sigma, type="C"):
    "Calculate BS option price for a call/put"
    d1 = (np.log(S/K) + (r + sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    try:
        if type == "C":
            price = S*norm.cdf(d1, 0, 1) - K*np.exp(-r*T)*norm.cdf(d2, 0, 1)
        elif type == "P":
            price = K*np.exp(-r*T)*norm.cdf(-d2, 0, 1) - S*norm.cdf(-d1, 0, 1)
        return price
    except:
        print("Please confirm all option parameters above!")

        
print("Option price is: ", round(blackscholes(r, S, K, T, sigma, type="C"), 2) )
```
