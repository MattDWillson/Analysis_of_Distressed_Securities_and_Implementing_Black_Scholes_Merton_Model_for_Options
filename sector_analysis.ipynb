```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
#Read Hilton price data CSV file

csv_path = Path("./Resources/HLT_Closing_Price_Data.csv")
HLT_Closing_Price_Data = pd.read_csv(csv_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
HLT_Closing_Price_Data.sort_index(inplace=True)
HLT_Closing_Price_Data.columns = ["HLT"]
HLT_Closing_Price_Data.head()
```

```python
#Read Marriott price data CSV file 

csv_path = Path("./Resources/MAR_Closing_Price_Data.csv")
MAR_Closing_Price_Data = pd.read_csv(csv_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
MAR_Closing_Price_Data.sort_index(inplace=True)
MAR_Closing_Price_Data.columns = ["MAR"]
MAR_Closing_Price_Data.head()
```

```python
#Read S&P price data CSV file

csv_path = Path("./Resources/INX_Closing_Price_Data.csv")
INX_Closing_Price_Data = pd.read_csv(csv_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
INX_Closing_Price_Data.sort_index(inplace=True)
INX_Closing_Price_Data.columns = ["S&P"]
INX_Closing_Price_Data.head()
```

```python
#Read 10 year treasury bill price data CSV file 

csv_path = Path("./Resources/TNX_Closing_Price_Data.csv")
TNX_Closing_Price_Data = pd.read_csv(csv_path, index_col="Date", infer_datetime_format=True, parse_dates=True)
TNX_Closing_Price_Data.sort_index(inplace=True)
TNX_Closing_Price_Data.columns = ["10 year Treasury Bill"]
TNX_Closing_Price_Data.head()
```

```python
#Calculate Hilton daily returns 

HLT_returns_df = HLT_Closing_Price_Data.pct_change()
HLT_returns_df.dropna(inplace=True)
HLT_returns_df.head()
```

```python
#Calculate Marriott daily returns

MAR_returns_df = MAR_Closing_Price_Data.pct_change()
MAR_returns_df.dropna(inplace=True)
MAR_returns_df.head()
```

```python
#Calculate S&P daily returns

INX_returns_df = INX_Closing_Price_Data.pct_change()
INX_returns_df.dropna(inplace=True)
INX_returns_df.head()
```

```python
#Calculate Ten Year Treasury Bill daily returns

TNX_returns_df = TNX_Closing_Price_Data.pct_change()
TNX_returns_df.dropna(inplace=True)
TNX_returns_df.head()
```
